### ***GPU PRICE SCRAPING***

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.newegg.ca/gigabyte-geforce-rtx-4090-gv-n4090gaming-oc-24gd/p/N82E16814932550'
results = requests.get(url)
print(results)
```

##### *We can now make it better by using beautifulsouop*

```python
doc = bs(results.text, 'html.parser')
print(doc.prettify())
```

##### *Now we want to find the price of the gpu in this link*


```python
# we just have to search for the dollar symbol itself
price = doc.find_all(text='$')
price
```

##### *Now we will find tha parent of the dollar symbol so we can get the tag of the dollar symbol and price attached to the dollar symbol*

```python
parent = price[0].parent
parent
```

```python
actual_price = parent.find('strong').string
actual_price 
```
