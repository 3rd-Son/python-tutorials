```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url ='https://www.worldometers.info/coronavirus/country/us/'

response = requests.get(url)
soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
rows = soup.find('table',{'id':"usa_table_countries_today"}).find('tbody').find_all('tr')
print(len(rows))
```

```python
rows_2 = soup.find('table',{'id':"usa_table_countries_today"}).find_all('tbody')[1].find_all('tr')
print(len(rows_2))
```
