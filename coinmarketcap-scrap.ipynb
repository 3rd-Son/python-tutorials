```python
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
url = 'https://coinmarketcap.com/'
results = requests.get(url).text
doc = bs(results, 'html.parser')
```

```python
print(doc.prettify())
```

##### *scrap the table body*

```python
tbody = doc.find('tbody')
tbody.prettify()
```

##### *Now to get the trs*

```python
trs = tbody.contents
trs
```

##### *We can see the next sibling with the .next_sibling() function*

```python
trs[0].next_sibling
```

##### *We can see the previous sibling by using the .previous_sibling function()*

```python
trs[2].previous_sibling
```

##### *We can list all the tr after the first tr by using the .next_siblings() function*

```python
list(trs[0].next_siblings)
```

##### *You can also find the name of the parent*

```python
trs[0].parent.name
```
